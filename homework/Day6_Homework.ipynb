### 1. [簡答題] 請問下列這三種方法有什麼不同？

```
print(a.sum())

```

````python
import numpy as np 

aNp = np.arange(10)
aNp
````

````python
print(sum(a))
print(sum(aNp))
````

### 解答
1. lists, tuples, and np 中 array 皆可以用 內建函數 `sum()` 使用, `a.sum() `與 `np.sum()` 所適用的是 np 函式庫中的函數, 其中 `a.sum()`使用的是 a 物件裡面的方法。

### 2. 請對一個 5x5 的隨機矩陣作正規化的操作。


````python
A = np.random.random((5, 5))
print(A)
row_sums = A.sum(axis=1)
A = A / row_sums[:, np.newaxis]
print(A)
````

### 3. 請建立一個長度等於 10 的正整數向量，並且將其中的最大值改成 -1。


````python
A = np.random.random((5))
print(A)
A = A / np.linalg.norm(A) * 10
A[np.argmax(A)] = -1
print(A)
````
