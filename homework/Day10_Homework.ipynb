### 1. [簡答題] 請問下列四種不同的 DataFrame 選取結果有什麼差異？

```
df.loc[ '2013-01-01', 'A'] 
df.loc[ '2013-01-01', ['A', 'B'] ] 
df.loc[ '2013-01-01':'2013-01-02', 'A' ] 
df.loc[ '2013-01-01':'2013-01-05', 'A':'C'] 
```





df.loc[ '2013-01-01', 'A'] : 從 '2013-01-01' 這一列中選取 'A' 欄位的值。

df.loc[ '2013-01-01', ['A', 'B'] ] : 從 '2013-01-01' 這一列中選取 'A' 和 'B' 兩個欄位的值。

df.loc[ '2013-01-01':'2013-01-02', 'A' ] : 從 '2013-01-01' 到 '2013-01-02' 這兩列中選取 'A' 欄位的值。

df.loc[ '2013-01-01':'2013-01-05', 'A':'C'] : 從 '2013-01-01' 到 '2013-01-05' 這五列中選取 'A' 到 'C' 三個欄位的值。

可以發現，透過 loc 可以進行行和列的範圍選取，也可以選取指定的欄位。另外，這些操作都是基於索引標籤進行的，因此不需要考慮行或列的位置。

### 2. 請根據提供的資料，選擇出下列的要求：

```
# - select the first 3 rows.
# - select the odd rows. (index = 1, 3, 5)
# - select the last 2 columns.
# - select the even columns. (index = 0, 2, 4)
```


````python
# 記得先 Import 正確的套件
import pandas as pd
import numpy as np
````

````python
df = pd.DataFrame(np.random.randn(6,3),columns=['col1','col2','col3'])
df
````

````python
# - select the first 3 rows.
# df.iloc[0:3]
df.iloc[:3]
````

````python
# - select the odd rows. (index = 1, 3, 5)
df.iloc[1::2]
````

````python
# - select the last 2 columns.
df.iloc[:, -2:]
````

````python
# - select the even columns. (index = 0, 2, 4)
df.iloc[:, ::2]
````

### 3. 請根據提供的資料，選擇出下列的要求：

```
# - 1. filtered by first column > 20?
# - 2. filtered by first column + second column > 50
# - 3. filtered by first column < 30 or second column > 30
# - 4. filtered by total sum of row > 100
```

````python
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
df
````

````python
# - 1. filtered by first column > 20?
df.loc[df.iloc[:, 0]>20]
````

````python
# - 2. filtered by first column + second column > 50
df.loc[df.iloc[:, 0] + df.iloc[:, 1] > 50]
````

````python
# - 3. filtered by first column < 30 or second column > 30
df.loc[(df.iloc[:, 1] > 30) & (df.iloc[:, 0] < 30)]
````

````python
# - 4. filtered by total sum of row > 100
# 計算每一列的總和
sums = df.sum(axis=1)

# 產生布林遮罩
mask = sums > 100
df.loc[mask]
````
