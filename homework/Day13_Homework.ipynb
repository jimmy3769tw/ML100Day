### 1. 請接下列資料依照指定規則做合併：

```
- 依照 fruit 欄位做合併
- 依照 index 索引做合併
```


````python
df1 = pd.DataFrame({
    'fruit': ['apple', 'banana', 'orange'] * 3,
    'weight': ['high', 'medium', 'low'] * 3,
    'price': np.random.randint(0, 15, 9)
})

df2 = pd.DataFrame({
    'fruit': ['apple', 'orange', 'pine'] * 2,
    'weight': ['high', 'low'] * 3,
    'price': np.random.randint(0, 15, 6)
})
````

````python
df1
````

````python
df2
````

````python
# - 依照 fruit 欄位做合併
pd.merge(df1, df2, on='fruit')
````

其中 weight_x、price_x 是 df1 中的欄位，weight_y、price_y 是 df2 中的欄位。因為 df2 中沒有包含 banana 這個水果，所以對應的欄位值為 NaN。


````python
# - 依照 index 索引做合併
# pd.concat([df1, df2], axis=1)
df1.join(df2, lsuffix='_df1', rsuffix='_df2')
````

這裡的 join() 方法和前面的範例程式碼差不多，只是這次在接收的參數中，除了傳入要加入的 DataFrame 之外，也傳入了 lsuffix 和 rsuffix 兩個參數，用來確定欄位名稱的後綴，避免兩個 DataFrame 中出現同名欄位後重複。
